```python
y = pd.DataFrame(df_def.numero_de_ingresos)
x = df_def.drop(['numero_de_ingresos'],axis=1)
```

```python
fecha_train = '2013-01-01'
x_train = x[x.index < fecha_train]
x_test = x[x.index >= fecha_train]
y_train = y[y.index < fecha_train]
y_test = y[y.index >= fecha_train]
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
```

#### catboost

```python
from catboost import CatBoostRegressor



model_cat = CatBoostRegressor(iterations=1000, learning_rate=0.1, depth=5, l2_leaf_reg=3)
model_cat.fit(x_train, y_train)

preds_cat = []
for index, row in x_test.iterrows():
    pred = model_cat.predict(row)
    preds_cat.append(pred)
    pred_series = pd.Series(pred, index=['numero_de_ingresos'], name=index)
    train_x = x_train.append(row)
    train_y = y_train.append(pred_series)
    model_cat.fit(x_train, y_train)
```

#### xgboost

```python
fecha_train = '2013-01-01'
x_train = x[x.index < fecha_train]
x_test = x[x.index >= fecha_train]
y_train = y[y.index < fecha_train]
y_test = y[y.index >= fecha_train]
```

```python
from xgboost import XGBRegressor
model= XGBRegressor()
param_grid = {'n_estimators': [10, 100,200],
    'max_depth': [5, 10,20],
    'learning_rate':[0.01,0.001,0.1,1]
}
cv = GridSearchCV(model, param_grid, cv = 4, verbose = True, n_jobs = -1)
cv.fit(x_train,y_train)

model_xgb = XGBRegressor(**cv.best_params_)
model_xgb.fit(x_train, y_train)

preds_xgb = []
for index, row in x_test.iterrows():
    pred = model_xgb.predict(np.array([row]))
    preds_xgb.append(pred[0])
    pred_series = pd.Series(pred, index=['numero_de_ingresos'], name=index)
    train_x = x_train.append(row)
    train_y = y_train.append(pred_series)
    model_xgb.fit(x_train, y_train)
```

```python
results = y_test
```

```python
results['preds_xgb'] = preds_xgb
results['preds_cat'] = preds_cat
```

```python
results
```

```python
px.line(results)
```

```python
import shap

explainer = shap.TreeExplainer(model_cat)
shap_values = explainer.shap_values(x_test)
shap.summary_plot(shap_values, x_test)

explainer = shap.TreeExplainer(model_xgb)
shap_values = explainer.shap_values(x_test)
shap.summary_plot(shap_values, x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error

rmse_cat, mae_cat, mape_cat = np.sqrt(mean_squared_error(results['numero_de_ingresos'], results['preds_cat'])), mean_absolute_error(results['numero_de_ingresos'], results['preds_cat']), mean_absolute_percentage_error(results['numero_de_ingresos'], results['preds_cat'])
rmse_xgb, mae_xgb, mape_xgb = np.sqrt(mean_squared_error(results['numero_de_ingresos'], results['preds_xgb'])), mean_absolute_error(results['numero_de_ingresos'], results['preds_xgb']), mean_absolute_percentage_error(results['numero_de_ingresos'], results['preds_xgb'])
```

```python
pd.DataFrame({'mse':[rmse_cat,rmse_xgb], 'mae':[mae_cat,mae_xgb], 'mape':[mape_cat,mape_xgb]}, index=['catboost','xgboost'])
```
